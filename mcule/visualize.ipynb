```python
import rdkit
from rdkit import Chem
from rdkit.Chem.Draw import IPythonConsole
```

```python
def load_molecules(smiles_file: str, num: int=20) -> list[rdkit.Chem.rdchem.Mol]:
    """
    View molecules from a given file containing SMILES strings.
    """    
    mols = list()
        
    supplier = Chem.SmilesMolSupplier(smiles_file, smilesColumn=0, nameColumn=0, titleLine=False, sanitize=True)
    for mol in supplier:
        try:
            mol = Chem.AddHs(mol)
            mols.append(mol)
        except Exception as e:
            pass
        
        if len(mols) >= num:
            break
        
    return mols
```

```python
mols = load_molecules("output/aziridine.smi")

Chem.Draw.MolsToGridImage(mols, molsPerRow=4, useSVG=True)
```
